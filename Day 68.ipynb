```python
import matplotlib.pyplot as plt
```

```python
shops = ["A","B","C","D","E","F"]              #For categories
sale = [40000,2300,45333,23412,3333,55555]
plt.bar(shops, sale,color = "blue", label = "sales 2025" )
plt.xlabel("Shop Data")
plt.ylabel("Prices")
plt.legend()
plt.show()
```

```python
shops = ["A","B","C","D","E","F"]              #For categories
sale = [40000,2300,45333,23412,3333,55555]
plt.barh(shops, sale,color = "orange", label = "sales 2025" )
plt.xlabel("Shop Data")
plt.ylabel("Prices")
plt.legend()
plt.show()
```

```python
season = ["South","East","West","North"]
value =[2000,1500,1300,1000]
plt.pie(value, labels = season, autopct="%1.1f%%", colors = ["gold", "lightblue", "blue","purple"])
plt.legend()
plt.title("Data by region")
plt.show()
```
